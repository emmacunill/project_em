```python

def clean_year(year):
    pattern_year = r'\d{4}'
    str_pattern = r'\d{4}\.'
    if re.match(str_pattern, str(year)):
        str_ = re.match(pattern_year, str(year)).group()
        return int(str_)
    else:
        return 'unknown'
    
'''To make it 4 digits and if not unknown and to turn them into ints'''
```

```python
def extract_age(x):

    '''first I transform manually the nan and the values with only 1 to 3 variations
      like teens, adult, 18 month.'''
    
    if pd.isna(x):
        return "unknown"
    if x == "teen" or x == "Teen" or x == "Teens":
        return "15"
    if x == "adult" or x == "(adult)" or x == "middle-aged":
        return "50"
    if x == "18 months":
        return "2"
    
    '''create a pattern to make the averege of the different ages of the values that give you 
    an age, then & ,or, to. Then applied the pattern, transformed the strings into ints to 
    get the average, get the items out of the list and transform them into string again.'''

    age_av = re.findall(r'(\d{1,2})\s*(&|or|to)\s*(\d{1,2})', str(x))
    
    if age_av:
        average_ages = [round((int(match[0]) + int(match[2])) / 2) for match in age_av]
        return str(average_ages[0])
    
    '''Also made a pattern to fins the values with text and numbers to catch 
    the 1 or 2 numbers to make the age.'''

    age_match = re.search(r'\d{1,2}', str(x))
    if age_match:
        return age_match.group()
    return "unknown"
```

```python
sharks_["age"] = sharks_["age"].apply(extract_age)
```

    ''' As we can see there are tons of activities and explanations, so I'm going to create a function to count key words and from that create the redex patterns to clean the column.'''

```python
def find_key_words(items):
    items = [str(i) for i in items]
    text = ' '.join(items)
    words = text.split()
    counter = Counter(x for x in words if len(x)>2)
    rkw= [word for word, count in counter.items() if count > 1]
    return rkw
```

```python
activities = set(sharks_.activity)
repeated_key_words = find_key_words(activities)
for word in repeated_key_words:
    print(word)
```

```python
def activity_categorize(input_str):


    if pd.isna(input_str):
        return "other"    
    if re.search(r'.*swim.*', input_str, re.IGNORECASE):
        return "swimming"
    if re.search(r'.*fishing.*', input_str, re.IGNORECASE) or re.search(r'.*hunting.*', input_str, re.IGNORECASE) or re.search(r'.*net.*', input_str, re.IGNORECASE):
        return "fishing"
    if re.search(r'.*surf.*', input_str, re.IGNORECASE) or re.search(r'.*board.*', input_str, re.IGNORECASE) or re.search(r'.*paddeling.*', input_str, re.IGNORECASE) or re.search(r'.*treading.*', input_str, re.IGNORECASE):
        return "surfing"
    if re.search(r'.*div.*', input_str, re.IGNORECASE):
        return "diving"
    if re.search(r'.*boat.*', input_str, re.IGNORECASE) or re.search(r'.*sink.*', input_str, re.IGNORECASE) or re.search(r'.*ship.*', input_str, re.IGNORECASE) or re.search(r'.*bar.*', input_str, re.IGNORECASE) or re.search(r'.*wreck.*', input_str, re.IGNORECASE) or re.search(r'.*raft.*', input_str, re.IGNORECASE)or re.search(r'.*yacht.*', input_str, re.IGNORECASE)or re.search(r'.*submarine.*', input_str, re.IGNORECASE):
        return "sailing"
    if re.search(r'.*air.*', input_str, re.IGNORECASE) or re.search(r'.*constellation.*', input_str, re.IGNORECASE) or re.search(r'.*parachute.*', input_str, re.IGNORECASE):
        return "air accidents"
    if re.search(r'.*cano.*', input_str, re.IGNORECASE) or re.search(r'.*kayak.*', input_str, re.IGNORECASE):
        return "non motor"
    if re.search(r'.*explosive.*', input_str, re.IGNORECASE) or re.search(r'.*anestesize.*', input_str, re.IGNORECASE) or re.search(r'.*pull.*', input_str, re.IGNORECASE) or re.search(r'.*investigate.*', input_str, re.IGNORECASE) or re.search(r'.*film.*', input_str, re.IGNORECASE) or re.search(r'.*photo.*', input_str, re.IGNORECASE) or re.search(r'.*defecat.*', input_str, re.IGNORECASE) or re.search(r'.*chumm.*', input_str, re.IGNORECASE) or re.search(r'.*wrang.*', input_str, re.IGNORECASE) or re.search(r'.*murder.*', input_str, re.IGNORECASE) or re.search(r'.*search.*', input_str, re.IGNORECASE):
        return "shark interacting"
    if re.search(r'.*natural.*', input_str, re.IGNORECASE) or re.search(r'.*hurricane.*', input_str, re.IGNORECASE):
        return "natural disasters"
    if re.search(r'.*bath.*', input_str, re.IGNORECASE) or re.search(r'.*float.*', input_str, re.IGNORECASE) or re.search(r'.*stand.*', input_str, re.IGNORECASE) or re.search(r'.*sit.*', input_str, re.IGNORECASE) or re.search(r'.*dangling.*', input_str, re.IGNORECASE) or re.search(r'.*crouching.*', input_str, re.IGNORECASE):
        return "bathing"
    return "other"

''' Function to categorize the activities: 
    - First I descard the nan because they are floats and it breaks the function.
    - Then with regex I input the patterns on how to categorize
    ps: literally crying that it worked, spent like 6 hours on that.'''
```
